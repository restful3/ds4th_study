## 피처 요약표 생성 함수

```python
import pandas as pd

def resumetable(df):
    print(f'데이터셋 형상: {df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입'])
    summary = summary.reset_index()
    summary = summary.rename(columns={'index': '피처'})
    summary['결측값 개수'] = df.isnull().sum().values # Series 는 기본적으로 인덱스(bin_0, bin_1 ..)있음, 값만 쓰기 위해 value
    summary['고윳값 개수'] = df.nunique().values
    summary['첫 번째 값'] = df.loc[0].values
    summary['두 번째 값'] = df.loc[1].values
    summary['세 번째 값'] = df.loc[2].values
    return summary    
```

# 메모리 절약을 위한 데이터 다운 캐스팅

## 데이터 다운캐스팅

```python
import pandas as pd

data_path = '/home/restful3/workspaces/study/ds4th_study/data/predict-future-sales/'

sales_train = pd.read_csv(data_path+'sales_train.csv')
shops = pd.read_csv(data_path+'shops.csv')
items = pd.read_csv(data_path+'items.csv')
item_categories = pd.read_csv(data_path+'item_categories.csv')
```

```python
train = sales_train.merge(shops, on='shop_id', how='left')
train = train.merge(items, on='item_id', how='left')
train = train.merge(item_categories, on='item_category_id', how='left')
```

```python
start_mem = train.memory_usage().sum()/1023**2
start_mem
```

- object : 그대로 둠
- bool : int8 로 변환
- 정수형 타입 또는 실수형 타입 중 소수점 첫째 자리에서 반올림한 수가 원래의 수와 같은 경우 : 해당 타입의 가장 작은 타입으로
    - 예 : 1.0, 10.0, 12.0 을 1, 10, 12로 변형
- 그외 : 실후형 타입 중 해당 데이터의 가장 작은 타입으로

```python
for col in train.columns:
    dtype_name = train[col].dtype.name
    if dtype_name == 'object':
        pass
    elif dtype_name == 'bool':
        train[col] = train[col].astype('int8')
    elif dtype_name.startswith('int') or (train[col].round() == train[col]).all():
        train[col] = pd.to_numeric(train[col], downcast='integer')
    else:
        train[col] = pd.to_numeric(train[col], downcast='float')
```

## 데이터 다운캐스팅 함수
- 주의 : object, bool, int, float 이외의 데이터 타입이 있다면 수정 필요

```python
def downcast(df, verbose=True):
    start_mem = train.memory_usage().sum()/1023**2

    for col in train.columns:
        dtype_name = train[col].dtype.name
        if dtype_name == 'object':
            pass
        elif dtype_name == 'bool':
            train[col] = train[col].astype('int8')
        elif dtype_name.startswith('int') or (train[col].round() == train[col]).all():
            train[col] = pd.to_numeric(train[col], downcast='integer')
        else:
            train[col] = pd.to_numeric(train[col], downcast='float')    
            
    end_mem = train.memory_usage().sum()/1023**2
    if verbose:
        print('{:.1f}압축됨'.format(100*(start_mem - end_mem) /  start_mem))
```

```python
downcast(train, True)
```
