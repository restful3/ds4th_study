# 데이터프레임 만들기

```python
import pandas as pd

# 비어 있는 데이터 프레임을 만들어 개별적으로 열 추가
df = pd.DataFrame()
df['Name'] = ['Song', 'Park', 'Lee']
df['Age'] = [22, 21, 19]
df['Driver'] = [True, False, False]
df
```

```python
# 데이터 프레임 객체를 만든 후 행 추가

new_person = pd.DataFrame({'Name':['Molly'], 'Age':[40],'Driver': [True]})
# new_person = pd.Series(['Molly',40,'Driver'], index=['Name', 'Age', 'Driver'] )
df = pd.concat([df, new_person], ignore_index=True, axis=0)
df
```

```python
# 넘파이 배열로 데이터 프레임 만들기

import numpy as np
data = [['Jack', 34, True], ['Steven', 22, False]]
matrix = np.array(data)
pd.DataFrame(matrix, columns=['Name', 'Age', 'Driver'])
```

```python
# 리스트로 만들기
data = [['Jack', 34, True], ['Steven', 22, False]]
pd.DataFrame(data, columns=['Name', 'Age', 'Driver'])
```

```python
# 딕셔너리로 만들기
data = {'Name':['Jack', 'Steven'],
       'Age':[34, 22],
       'Driver':[True, False]}

pd.DataFrame(data)
```

```python
# 샘플마다 딕셔너리 
data = [
    {'Name':'Jack', 'Age':34, 'Driver': True},
    {'Name':'Steven', 'Age':22, 'Driver': False}    
]
pd.DataFrame(data, index=[0, 1])
```

# 모든 열 원소에 함수 적용하기

```python
import pandas as pd

data_path = './datasets/titanic.csv'
df = pd.read_csv(data_path)
```

```python
def uppercase(x):
    return x.upper()

df['Name'].apply(uppercase)[:3]
```

```python
df['Survived'].map({1:'Alive', 0:'Dead'})[:3]
```

```python
df['Age'].apply(lambda x, age : x<age , age=30)[:5]
```

```python
# 숫자 데이터만 선택하여 새로운 DataFrame을 만듦
numeric_df = df.select_dtypes(include=[np.number])

# 각 열의 최댓값을 구함
numeric_df.apply(lambda x : max(x))
```

```python
def truncate_string(x):
    if type(x) == str:
        return x[:10]
    else:
        return x
```

```python
df.applymap(truncate_string)[:4]
```

# 데이터프레임 병합하기

```python
import pandas as pd

employee_data = {'employee_id':[1, 2, 3, 4], 
                'name':['Amy Jones', 'Allon Keys', 'Alice Bees', 'Tim Horton']}
df_employee = pd.DataFrame(employee_data)
df_employee
```

```python
sales_data = {'employee_id':[3, 4, 5, 6], 
            'total_sales':[2345,534,567567,34334]}
df_sales = pd.DataFrame(sales_data)
df_sales
```

```python
pd.merge(df_employee, df_sales, on='employee_id', how='outer')
```

```python
pd.merge(df_employee, df_sales, on='employee_id', how='left')
```

```python
pd.merge(df_employee, df_sales, on='employee_id', how='inner')
```
