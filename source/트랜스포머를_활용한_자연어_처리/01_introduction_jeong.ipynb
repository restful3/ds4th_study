## 트랜스포머 애플리케이션 둘러보기
고객의 불평불만에 대해, 전이모델을 사용하여 분류하고 분석하고 요약하고, 답글을 생성해보자

```python
text = """Dear Amazon, last week I ordered an Optimus Prime action figure \
from your online store in Germany. Unfortunately, when I opened the package, \
I discovered to my horror that I had been sent an action figure of Megatron \
instead! As a lifelong enemy of the Decepticons, I hope you can understand my \
dilemma. To resolve the issue, I demand an exchange of Megatron for the \
Optimus Prime figure I ordered. Enclosed are copies of my records concerning \
this purchase. I expect to hear from you soon. Sincerely, Bumblebee."""
```

### 텍스트 분류

```python
# 모델 가져와서 classifier라고 담아
from transformers import pipeline

classifier = pipeline("text-classification")
```

```python
# 모델에 텍스트 넣어서 결과를 볼까
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

꽤나 부정적인걸 알수 있어

### 개체명 인식
- 고객 피드백의 감성을 예측하는 것은 좋은 분석의 출발점이 되지만, 피드백이 특정 제품과 서비스 중 무엇에 대한 것인지 알고 싶을 때 많음
    - 제품, 장소, 사람 과 같은 실제 객체를 개체명(named entity)라고 하며 이런 개체명을 텍스트에서 추출하는 작업을 개체명인식(Named Entity Recognition)이라 한다.

```python
# 해당 파이프라인을 로드하고 고객 리뷰를 전달하여 NER을 적용함
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

> - 모든 개체명을 감지하고 ORG(조직), LOC(위치), PER(사람) 같은 카테고리에 할당
> - 모델 예측에 따라 단어를 그룹화하기 위해 aggregation_strategy라는 매개변수 사용
> - 점수는 모델의 개체명을 얼마나 확신하는지를 나타냄

### 질문 답변
- 텍스트 구절과 함께 답을 얻고 싶은 질문을 모델에 전달하고 모델은 답변 텍스트를 변환.

```python
# 질문대답 모델가져와서 질문해
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

질문에 해당하는 답변을 가져오는걸 볼수있지

### 요약
- 관련 사실이 모두 포함된 간단한 버전을 생성. 모델이 논리적인 텍스트를 생성해야 하므로 이전 문제보다 훨씬 복잡한 작업.
    - 요약 파이프라인을 활용

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=56, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

### 번역
- 번역 파이프라인을 사용하여 영어텍스트를 독일어로 번역.
    - 허깅페이스허브에는 수천 개 언어 쌍에 대한 모델 있음

```python
translator = pipeline("translation_en_to_de",
                      model="Helsinki-NLP/opus-mt-en-de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text'])
```

### 텍스트 생성
- 자동 완성 기능으로 고객 피드백에 빠르게 응답하고 싶음. 텍스트 생성모델 활용

```python
from transformers import set_seed
set_seed(42) # 동일 결과를 재현하기 위해 지정
```

```python
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text + " {with giving clear solution to solve the problem as a good service},"+"\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```

보다 적절한 답변을 이끌기 위해 프롬프트에 {}를 통해 속성을 추가

## 트랜스포머의 주요 도전 과제
> 1. 언어
> - 주요언어가 영어지만 데이터 부족. 다중언어 트랜스포머 및 제로샷교차언어전이 수행은 4장에 나옴

<br>

> 2. 데이터 가용성
> - 전이학습을 통해 훈련데이터가 많이 줄어도, 사람이 작업 수행하는데 필요한 양은 여전히 많음
>     - 데이터가 없거나 소량은 경우 9장에서 학습

<br>

> 3. 긴 문서 처리하기
> - 셀프어텐션은 문단정도 될 때 잘 작동. 논문처럼 길어지면 비용 많이 듬.
>     - 이를 완화하는 방법 11장

<br>

> 4. 불투명성
> - 트랜스포머모델은 주로 인터넷의 텍스트데이터를 통해 사전 훈련하므로, 데이터의 편향이 데이터에 고스란히 전이
