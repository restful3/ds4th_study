## 랭체인의 기능
- Langchain이란? 언어모델 구동되는 애플리케이션을 개발하기 위한 프레임워크
  - 데이터 소스에 기반하여 학습하도록 연결
  - 언어모델이 환경과 상호작용하도록 작동 : 인터넷연결/검색 등 타 서비스 연동
- ChatGPT있는데 Langchain은 왜 써야하지?
  - chatGPT만으로 최근정보나 특정분야의 정보를 알수 없음, 할루셔네이션 발생
  - chatGPT는 각 4096(GPT3.5), 8192(GPT4) 토큰이라는 입력제한이 있음
- ChatGPT 단점 보안 방향
  - Finetuning : 딥러닝의 weight 조정하여 원하는 용도 모델로 업데이트
    - 전이학습하는 방식<br>
  - N-shot Learning : 0~n개의 출력예시를 제시하여 용도에 맞는 출력을 하도록 조정
    - 낮은 수준의 조정<br>
  - In-context Learning : 문맥을 제시하고, 이 문맥 기반으로 모델이 출력되도록 조정
    - Langchain 활용  <br>
- LangChain이 어떻게 chatGPT의 문제를 해결하는데?
  - 정보부족 해결
    > Vectorstore 기반 정보탐색 | Agent활용한 검색 결합
  - 토큰제한 해결
    > Textsplit 활용한 문서분할
  - 환각현상 해결
    > 주어진문서에만 답하도록 Prompt입력
- LangChain 구조
  - LLM : 초거대 언어모델로 생성모델의 엔진과 같은 역할
    > GPT, PLLM, LLAMA, ... 등 연결
  - Prompt : 언어모델에게 지시하는 명령문
    > Prompt templates, Chat Prompt Template, Example Selectors, Output Parsers
  - Index : LLM이 문서를 쉽게 탐색할 수 있도록 구조화하는 모듈
    > Document Loader, Text Splitters, Vertorstores, Retrivers,...
  - Memory : 채팅이력을 기억하여, 이어서 대화가 가능하도록 하는 모듈
    > ConversationBufferMemory, Entity Memory, Conversation Knowledge Graph Memory,...
  - Chain : LLM 사슬을 형성하여 연속적인 LLM 호출이 가능하도록 하는 핵심구성요소
    > LLM Chain, Question Answering, Summarization, Retrival Question/Answering,...
  - Agents :기존 Prompt Template으로 수행할 수 없는 작업을 가능토록 하는 모듈
    > Custom Agent, Custom MultiAction Agent, Conversation Agent,...
