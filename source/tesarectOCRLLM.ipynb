```python
'''
이미지를 그레이스케일하여 (인식률을 높이고) Tesseract로 OCR하는 코드입니다. 
'''
import pytesseract
from PIL import Image
import cv2
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract'

def ocr(img_path):
    img_cv = cv2.imread(img_path+'.png')
    img_rgb=cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
    # img_rgb=cv2.cvtColor(img_cv, cv2.IMREAD_GRAYSCALE)
    config = ('-l kor+eng --oem 3 --psm 4')
    img_pytesseract_en = pytesseract.image_to_string(img_rgb,config=config)
    return print(img_pytesseract_en)
```

```python
# pdf2image 라이브러리를 import합니다. PDF 파일을 이미지로 변환하는 데 사용됩니다.
import pdf2image

# pytesseract 라이브러리를 import합니다. 이미지에서 텍스트를 추출하는 데 사용됩니다.
import pytesseract

try:
    # PIL (Python Imaging Library) 라이브러리를 import합니다. 이미지 처리에 사용됩니다.
    from PIL import Image
except ImportError:
    # PIL 라이브러리를 import할 수 없는 경우 대신 Image 모듈을 import합니다.
    import Image

# PDF 파일을 이미지로 변환하는 함수입니다.
def pdf_to_img(pdf_file):
    return pdf2image.convert_from_path(pdf_file)

# 이미지에서 텍스트를 추출하는 함수입니다.
def ocr(img_path):
    # OCR 엔진의 설정을 지정합니다. 한국어와 영어를 인식하고, 특정 OCR 엔진과 페이지 세그멘테이션 모드를 사용합니다.
    config = ('-l kor+eng --oem 3 --psm 4')
    # pytesseract 라이브러리를 사용하여 이미지에서 텍스트를 추출합니다.
    img_pytesseract= pytesseract.image_to_string(img_path,config=config)
    return img_pytesseract

# PDF 파일의 각 페이지에서 텍스트를 추출하는 함수입니다.
def print_pages(pdf_file):
    # PDF 파일을 이미지로 변환합니다.
    images = pdf_to_img(pdf_file)
    tmp = {}
    # 각 페이지에 대해 반복합니다.
    for pg, img in enumerate(images):
        # 페이지 번호를 키로 하고, 해당 페이지의 텍스트를 값으로 하는 딕셔너리 항목을 생성합니다.
        tmp[f'{pg}페이지'] = ocr(img)
    return tmp

# print_pages 함수를 호출하여 PDF 파일의 텍스트를 추출합니다.
test = print_pages('D:/drive/SelfStudy/Books/퀀트/ch4.pdf')
# 추출된 텍스트를 출력합니다.
for i in test.items() :
    str_ = ''.join(str(i))
```
