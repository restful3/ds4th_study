# Document Loader
- Document Loaders는 다양한 형식의 문서를 불러오고
- 이를 langchain 에 결합하기 쉬운 텍스트 형태로 변환
- pdf, word, ppt, xlsx, csv 등 모든 형식의 문서를 기반으로 LLM구동 가능

# URL Document Loader
- 웹에 기록된 글도 텍스트 형식으로 읽어와 LLM에 활용 
- WebBasedLoader와 UnstructuredURLLoader

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://n.news.naver.com/mnews/article/421/0007282780?sid=102")
data = loader.load()
print(data[0].page_content)
```

```python
from langchain.document_loaders import UnstructuredURLLoader

urls = [
    "https://n.news.naver.com/mnews/article/421/0007282780?sid=102",
    "https://n.news.naver.com/mnews/article/214/0001323978?sid=100"
]

loader = UnstructuredURLLoader(urls = urls)
data = loader.load()
print(data)
```
