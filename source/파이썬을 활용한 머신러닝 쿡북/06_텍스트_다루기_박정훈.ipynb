6.1 텍스트 정제하기

"______i am a boy."을 "X XX X XXX"로 만들기
---

- 공백 제거
- 마침표 제거
- 대문자/소문자
- 문자를 모두(일부) X로 만들기

```python
text = ["  i am a boy.",
        " you are a girl.",
        "   i am hungry.    "]
text
```

```python
# strip : 공백 문자를 제거
strip_whitespace = [string.strip() for string in text]
print(strip_whitespace)
print("\n")



string = "---Hello, World!---"

'''특정 문자 제거'''
removed = string.strip('-')
print("특정문자 제거")
print(removed)
print("\n")
'''왼쪽에 있는 '-' 제거'''
left_stripped = string.lstrip('-')
print("왼쪽에 있는 특정 문자 제거")
print(left_stripped)
```

```python
'''마침표 제거'''
# replace : 문자를 대체
remove_periods = [string.replace(".","") for string in strip_whitespace]
remove_periods
```

```python
'''대문자로 만들기 (함수 활용)'''
# upper : 대문자로 변환

def capitalizer(string: str) -> str:
  return string.upper()

[capitalizer(string) for string in remove_periods]
```

re 라이브러리

re.search(pattern, string):
문자열에서 특정 패턴에 찾아서 반환

re.match(pattern, string): 문자열의 시작부분이 특정 패턴과 동일한지를 확인

re.findall(pattern, string): 문자열 내에서 패턴에 맞는 모든 문자열을 찾아서 리스트로 반환

re.finditer(pattern, string): 각 부분 문자열에 대한 정보를 얻을 수 있습니다.

re.sub(pattern, replacement, string): 문자열에서 다른 문자열로 치환

re.compile(pattern): 정규 표현식 패턴을 컴파일하여 재사용 가능한 객체를 생성합니다.
pattern: 정규 표현식을 나타내는 문자열입니다. 패턴 안에 특정한 규칙을 따라 검색할 문자열의 구조를 표현합니다.
replacement: re.sub()에서 사용되며, 패턴과 일치하는 부분을 치환할 문자열을 지정합니다.

```python
import re

text = "apple banana cherry date"
pattern = r"\b\w+\b"

matches = re.finditer(pattern, text)

for match in matches:
    start = match.start()
    end = match.end()
    matched_string = match.group()
    print(f"Match: {matched_string} at positions {start}-{end-1}")
```

```python
import re

pattern = r"\d{3}-\d{2}-\d{4}"  # 형식: 숫자-숫자-숫자

regex = re.compile(pattern)
text = "주민등록번호: 123-45-6789, 987-65-4321"
matches = regex.findall(text)

for match in matches:
    print(match)
print('\n')
for match in matches:
    print("주민등록번호:", match)
```

```python
''' a,b를 X로 치환 '''

import re

def replace_X(string:str) -> str:
    return re.sub(r"[a-b, A-B]","X",string)

print([replace_X(string) for string in remove_periods])
```

```python
from bs4 import BeautifulSoup

''' HTML 문서의 문자열 형태를 html 변수에 저장 '''
html = """
<div class='full_name'><span style='font-weight:bold'>masego</span>Azra</div>"""

# BeautifulSoup 객체를 생성하고, 두 개의 인자를 전달해줌
# 1. 파싱할 HTML 문자열(html 변수)이며,
# 2. 파서로 사용할 패키지 지정
# 본 교재에서는 "lxml"파서를 통해서 HTML을 파싱함
# lxml은 XML과 HTML 문서를 파싱/처리를 위한 고성능 라이브러리
soup = BeautifulSoup(html,"lxml")

''' find() 메서드를 사용하여 HTML 문서 내에서 특정 요소를 찾음 '''
soup.find("div",{"class":"full_name"}).text

''' Output '''
# span 요소의 텍스트("masego")와 div 요소의 텍스트("Azra")가 결합됨
```

# 6.7 품사태깅하기

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag
from nltk import word_tokenize

text_data = "Chris loved outdoor running"

text_tagged = pos_tag(word_tokenize(text_data))

print("토큰에 대한 품사")
print(text_tagged)
print("\n")


print("특정 품사(명사) 찾기")
[word for word, tag in text_tagged if tag in ['NN','NNS','NNP','NNPS'] ] # 명사 찾기
```

NN: 명사 (Noun)

NNS: 복수형 명사 (Noun, plural)

NNP: 고유 명사 (Proper noun, singular)

NNPS: 복수형 고유 명사 (Proper noun, plural)

VB: 동사 (Verb, base form)

VBD: 동사, 과거형 (Verb, past tense)

VBG: 동사, 현재 분사형 (Verb, gerund or present participle)

VBN: 동사, 과거 분사형 (Verb, past participle)

VBP: 동사, 현재형 (Verb, non-3rd person singular present)

VBZ: 동사, 3인칭 단수 현재형 (Verb, 3rd person singular present)

JJ: 형용사 (Adjective)

JJR: 비교급 형용사 (Adjective, comparative)

JJS: 최상급 형용사 (Adjective, superlative)

RB: 부사 (Adverb)

RBR: 비교급 부사 (Adverb, comparative)

RBS: 최상급 부사 (Adverb, superlative)

IN: 전치사 (Preposition or subordinating conjunction)

DT: 관형사 (Determiner)

PRP: 인칭 대명사 (Personal pronoun)

CC: 접속사 (Coordinating conjunction)

CD: 숫자 (Cardinal number)

```python
''' 품사를 특성으로 표현하기 '''
from sklearn.preprocessing import MultiLabelBinarizer
tweets = ["I am eating a burrito for breakfast",
          "Political science is an amazing field",
          "San Francisco is an awesome city"]

tagged_tweets = []

# DT 한정사, IN 전치사, JJ 형용사, NN 명사, NNP 고유명사, PRP (Personal Pronoun): 인칭 대명사, VBG 동명사, VBP 현재형 동사, VBZ 3인칭 단수 현재형 동사

''' 명사가 있을 경우 '''
for tweet in tweets:
    tweet_tag = nltk.pos_tag(word_tokenize(tweet))
    tagged_tweets.append([tag for word, tag in tweet_tag])

print("tagged_tweets")
print(tagged_tweets)
print("\n")

one_hot_multi = MultiLabelBinarizer()
one_hot_multi.fit_transform(tagged_tweets)
print("품사")
print(one_hot_multi.classes_)
print("\n")
print("원핫인코딩을 사용하여 품사를 특성으로 변환함")
print(one_hot_multi.fit_transform(tagged_tweets))
```

본인이 만든 테그 모델의 정확도를 확인하기 위한 방법론

```python
import nltk
nltk.download('brown')
```

```python
from nltk.corpus import brown
from nltk.tag import UnigramTagger
from nltk.tag import BigramTagger
from nltk.tag import TrigramTagger

'''브라운 코퍼스에서 텍스트를 추출한 후 문장으로 나눔'''
# 총 데이터수는 4623개
sentences = brown.tagged_sents(categories='news')

'''훈련/테스트용을 나눠줌'''
train = sentences[:4000]
test = sentences[4000:]

'''백오프 태그 객체'''
# UnigramTagger, BigramTagger, TrigramTagger은 품사태깅 모델
# UnigramTagger : 단어의 품사를 예측하기 위해 해당 단어만을 고려하고, 해당 단어가 가장 자주 등장하는 품사로 예측
# BigramTagger : 해당 단어와 바로 이전 단어를 함께 고려
# TrigramTagger : 해당 단어와 앞/뒤 두 단어를 함께 고려

unigram = UnigramTagger(train)
bigram = BigramTagger(train, backoff=unigram)
# backoff=unigram를 사용하는 이유
# BigramTagger는 이전 단어의 정보로부터 예측할 수 없는 단어의 품사를 UnigramTagger를 통해 예측하게 됨
trigram = TrigramTagger(train, backoff=bigram)

'''정확도 확인'''
trigram.evaluate(test)
```

6.8 텍스트를 BoW로 인코딩하기

(특정 단어의 등장 횟수를 나타내는 특성)
- 빈도수를 활용하여 중요도 도출 가능
- 방대한 양의 텍스트의 경우 특성행렬로 저장공간 축소 가능

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

text_data = np.array(['I love Brazil Brazil ! ',
                      'Sweden is best Brazil',
                      'Germany beats both'])

''' BoW 특성행렬을 만듬'''
count = CountVectorizer()
bag_of_words = count.fit_transform(text_data)

print('특성이름')
print(count.get_feature_names_out())
print('단어의 카운트 행렬')
print(bag_of_words.toarray())
# print(bag_of_words)


print("*"*70)
a = count.get_feature_names_out()
b = bag_of_words.toarray()
c = pd.DataFrame(b, columns=a)
display(c)
```

```python
# CounVtectorizer로 희소행렬의 방대한 용량을 줄일 수 있음

# 0 0 0 2 0 0 0 0
# (2,0,0)

# ngram : min_n, max_n를 기입
# stop_words : 내장된 리스트를 통해서 유용하지 않은 단어 제거 가능
# vocabulary : 분석대상을 지정 (단어나 구)
count_2gram = CountVectorizer(ngram_range=(1,2),
                              stop_words="english",
                              vocabulary=['brazil'])
bag = count_2gram.fit_transform(text_data)

print("특성행렬 확인")
print(bag.toarray())
print("1-그램, 2-그램 확인")
print(count_2gram.vocabulary_)
```

# 6.9 단어 중요도에 가중치 부여하기

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

text_data = np.array(['I love Brazil. Brazil!',
                      'Sweden is best ',
                      'Germany beats both '])

''' 특성행렬 만들기 '''
# 문서 집합을 TF-IDF(Term Frequency-Inverse Document Frequency)로 수치화
# TF : 문서 내 단어의 등장 빈도 (특정 문서에 등장하는 단어는 중요도가 높다)
# IDF : 등장빈도의 역수 (문서 전체에 등장하는 단어는 중요도가 낮다)
# TF-IDF : TF와 IDF의 곱

tfidf = TfidfVectorizer()
feature_matrix = tfidf.fit_transform(text_data)

print('*****feature_names*****')
print(tfidf.get_feature_names_out())
print("\n")

print('*****feature_matrix, array*****')
display(feature_matrix.toarray())
```
